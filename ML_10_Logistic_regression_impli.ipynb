# BINARY Class Classification Problem (logistic regression)

```python
df=df[df.target!=2]
```

```python
df.target.unique()
```

```python
df
```

```python
#NOw do Data preperation, Eda
```

```python
#divide into x and y
```

```python
x=df.iloc[:,:-1]       #all row all columns excepting last
y=df.iloc[:,-1]      #all the row and last column
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=123)
```

```python
x_train.shape, y_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_model = LogisticRegression()
```

```python
log_model
```

```python
log_model.fit(x_train,y_train)
```

```python
y_pred = log_model.predict(x_test)
```

```python
y_pred
```

```python
log_model.predict_proba(x_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

##Roc AUC

```python
from sklearn.metrics import roc_auc_score , roc_curve,auc
```

```python
y_pred_proba= log_model.predict_proba(x_test)[:,1]
```

```python
fpr, tpr , thresholds=roc_curve(y_pred,y_pred_proba)
```

```python
thresholds
```

```python
tpr
```

```python
fpr
```

```python
roc_auc=auc(fpr,tpr)
```

```python
roc_auc
```

```python
plt.figure(figsize=(8,8))
plt.plot(fpr,tpr,color='pink',linewidth=3,label='roc_curve(area=%0.2f)'%roc_auc) #plot roc_auc
plt.plot([0,1],[0,1],color='navy',linewidth=2,linestyle='--')#plot the diagonal line representing random guessing
plt.xlim([0.0,1.0])            #set x-axis limits
plt.ylim([0.0,1.05])           #set y-axis limits
plt.xlabel('False Positive Rate') #set  axis label
plt.ylabel('True Positive Rate')  # set y axis label
plt.title("Reciever Operating charcterristics(ROC) Curve") #Set the plot title
plt.legend(loc='lower right') # add legend to the plot
plt.show()
```
