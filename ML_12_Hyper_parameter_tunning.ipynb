# Hyperparameter Tunning

```python
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
x,y=make_classification(n_samples=1000,n_features=10,n_informative=5,n_redundant=5)

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=234)
```

```python
#grid search cv
from sklearn.model_selection import GridSearchCV
paransis={'penalty':('l1','l2','elasticnet'),'C':[1,2,3,4,5,10,20,30,40]}

classifier=LogisticRegression()
```

```python
clf=GridSearchCV(classifier,param_grid=paransis,cv=5,verbose=3)
```

```python
clf.fit(x_train,y_train)
```

```python
clf.best_score_
```

```python
clf.best_params_
```

```python
LogisticRegression(C=1,penalty="l2")
```

***Randmised search cv***

```python
from sklearn.model_selection import RandomizedSearchCV
clt=RandomizedSearchCV(classifier,param_distributions=paransis,cv=5,n_iter=10,verbose=2)
clt.fit(x_train,y_train)
```

```python
clt.best_score_
```
